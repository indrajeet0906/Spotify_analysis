```python
import matplotlib.pyplot as plt
```

```python
most_popular=df_tracks.query('popularity>90',inplace = False).sort_values('popularity',ascending= False)
most_popular[:10]
```

```python
df_tracks.set_index('release_date',inplace = False)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x:round(x/1000))
df_tracks.drop("duration_ms", inplace = True , axis=1)
```

```python
df_tracks.duration.head()


```

```python
sorted_df = df_tracks[['name', 'artists', 'popularity']].sort_values(by='popularity', ascending=False)
sorted_df.head()

```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
hist_spotify = df_tracks[['duration','popularity', 'danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']]

# Set the figure size and layout
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15, 20))
fig.subplots_adjust(hspace=0.5)

# Plot histograms for each variable in the dataframe
for ax, col in zip(axes.flatten(), hist_spotify.columns):
    ax.hist(hist_spotify[col], bins=30, color='Cyan', edgecolor='black', alpha=0.7, histtype='bar')
    ax.set_title(col, fontsize=14)
    ax.set_xlabel(col, fontsize=12)
    ax.set_ylabel('Frequency', fontsize=12)

# Display the plot
plt.suptitle('Distribution of Spotify Audio Features', fontsize=20)
plt.show()
```

```python
corr_df = df_tracks.drop(["key", "mode", "explicit"], axis=1).corr(method="pearson", numeric_only=True)

plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y="loudness", x="energy", color="y", line_kws={"color":"red"})
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y= "popularity", x = "acousticness", color = "b", line_kws={"color":"red"}).set(title="Popularity vs Acousticness")
```

```python
df_tracks['release_date'] = pd.to_datetime(df_tracks['release_date'], errors='coerce')
df_tracks['year'] = df_tracks['release_date'].dt.year

plt.figure(figsize=(12, 6))
sns.histplot(data=df_tracks, x='year', discrete=True, bins=30, kde=False)
plt.title('Number of Songs per Year')
plt.xlabel('Year')
plt.ylabel('Number of Songs')
plt.show()
```

```python
total_dr = df_tracks['duration']
years = df_tracks['year']

fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)

sns.barplot(x=years, y=total_dr, ax=ax, errorbar=None)
ax.set_title("Years vs Duration")
plt.xticks(rotation=90)

plt.show()
```

```python
total_dr=df_tracks['duration']
sns.set_style(style="whitegrid")
plt.figure(figsize = (10,5))
sns.lineplot(x=years, y=total_dr).set(title = "Year vs Duration")
plt.xlabel('Dates')
plt.ylabel('Duration')
plt.xticks(rotation=60)
plt.show()
```

```python
df_genre=pd.read_csv("spotifyfeatures.csv")
```

```python
df_genre.head()
```

```python
numerical_columns = df_genre.select_dtypes(include='number').columns
num_rows = (len(numerical_columns) + 1) // 2
num_cols = 2
fig, axes = plt.subplots(num_rows, num_cols, figsize=(7, 10))
axes = axes.flatten()
for i, column in enumerate(numerical_columns):
    ax = axes[i]
    df_genre[column].hist(ax=ax, bins=20)
    ax.set_title(column)
    ax.set_xlabel('Value')
    ax.set_ylabel('Frequency')
for i in range(len(numerical_columns), num_rows * num_cols):
    axes[i].axis('off')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(11,6))
plt.title("Duration of the Songs across Different Genres")
sns.color_palette("rocket", as_cmap = True)
sns.barplot (y='genre', x='duration_ms', data=df_genre)
plt.xlabel("Duration in milliseconds")
plt.ylabel("Genres")
```

```python
plt.figure(figsize = (5,4))
plt.title("top 5 Genre")
top_5 = df_genre.sort_values('popularity', ascending = False).head(10)
sns.barplot(data = top_5, x = 'popularity', y = 'genre')
plt.show()
```

```python
plt.figure(figsize = (5,4))
plt.title("Bottom 5 Genre")
bottom_5 = df_genre.sort_values('popularity', ascending = True).tail(10)
sns.barplot(data = bottom_5, x = 'popularity', y = 'genre')
plt.show()
```

```python
avg_popularity_by_genre = df_genre.groupby('genre')['popularity'].mean().sort_values(ascending=False)
fig, ax = plt.subplots(figsize=(10, 6))  
sns.barplot(x=avg_popularity_by_genre.index, y=avg_popularity_by_genre.values, palette="rocket")
plt.xlabel("Genre") 
plt.ylabel("Average Popularity")  
plt.title("Average Popularity by Genre")
plt.xticks(range(len(avg_popularity_by_genre.index)), rotation=90)  
plt.tight_layout()
plt.show()
```

```python
sns.set_style(style="darkgrid")
plt.figure(figsize=(10,5))
famous = df_genre.sort_values("popularity", ascending=False)
sns.barplot(y='genre',x='popularity', data=famous).set(title="Top 5 Geners based on Popularity")
```
